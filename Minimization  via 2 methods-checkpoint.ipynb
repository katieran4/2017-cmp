```python
#11/1 HW part a) Write a code from scratch to minimize the given function,not MC method

import numpy as np
import matplotlib.pyplot as plt

#Define the function
def f(x): #simple function with 1 minimum
    return x[0]**2/2 + x[1]**2/3 - x[0]*x[1]/4 

#define the derivative
def deriv(f, xy, d=0.01):
    x, y = xy[0], xy[1]
    fx = (f([x+d/2,y])-f([x-d/2,y]))/d
    fy = (f([x,y+d/2])-f([x,y-d/2]))/d
    #print(fx,fy)
    return np.array([fx,fy])

#Define initial values
xmin, xmax= -14,14
ymin, ymax= -14,14

#define random starting points
def start(xmin, xmax, ymin, ymax):
    x0 = xmin+np.random.random()*(xmax-xmin)
    y0 = ymin+np.random.random()*(ymax-ymin)
    return [x0, y0]

#Define a grid space in which to build the graphic
nx = np.linspace(xmin,xmax,400)
ny = np.linspace(ymin,ymax,400)

x,y = np.meshgrid(nx,ny)

#define the minimization function
def minimize(f,x0, N=1000):
    xnow = x0
    converged = False
    xhist = []
    xhist.append(xnow)
    dx = 0.01

    for i in range(N):
        
        dfnow = deriv(f, xnow) 
        xnext = xnow - dfnow*dx
        
        #print("step:    ", f(x_now), f(x_next))
        if f(xnext)>f(xnow):
            converged = True
            break
        else:
            xnow = xnext
            xhist.append(xnow)

    return converged, np.array(xhist), f(xnow)


[x0, y0] = start(xmin, xmax, ymin, ymax)
converged, xhist, fmin = minimize(f, [x0,y0])
x,y = np.meshgrid(nx,ny)
z = f([x, y])
fig = plt.figure()
levels = np.arange(np.min(z), np.max(z), 0.3)
plt.contour(x,y,z, levels=levels)

plt.plot(xhist[:,0], xhist[:,1], 'ro-')
plt.show()
print('results of fmin: ', fmin, '  number of iterations:  ', len(xhist))
```
