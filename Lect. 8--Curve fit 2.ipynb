```python

import matplotlib.pyplot as plt
import numpy as np

x = []
y = []

for i in np.loadtxt('years.txt'):
    x.append(i)

for j in np.loadtxt('US.txt'):
    y.append(j)


#fit
fit = np.polyfit(x, y, 4)
y_p = np.polyval(fit, x)


plt.plot(x, y, "o", label = 'Actual',color="green")
plt.plot(x, y_p, label = 'Trend', linewidth = 3.0, color = "darkorchid")
plt.xlabel("Year")
plt.ylabel("Population")
plt.xlim([1960, 2016])
plt.legend()
plt.show()

```
