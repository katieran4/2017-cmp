```python
#Generate plot

import matplotlib.pyplot as plt
import numpy as np
from random import random


fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')

x1 = np.linspace(0,2,100)
y1 = np.sqrt(1-(x1-1)**2)
ax1.plot(x1,y1,'k')  #upper function

y2 = 2-np.sqrt(4-x1**2)
ax1.plot(x1,y2,'k')  #lower function

ax1.set_xlim(0,2)
ax1.set_ylim(0,2)
ax1.fill_between(x1, y1, y2, where=y1>y2, facecolor='green')

plt.show()


#Calculate the enlosed area using monte carlo

def lower(x):
    return np.sqrt(1-(x-1)**2)

def upper(x):
     return 2-np.sqrt(4-x**2) 
    

def MC(N):
    count = 0
    for i in range (N):
            x=2*random()
            z=2*random()
            if lower(x)>z and z>upper(x):
                count += 1
    I= 4* count/N
    return I

print(MC(100))
print(MC(1000))
print(MC(10000))
print(MC(100000))
```
