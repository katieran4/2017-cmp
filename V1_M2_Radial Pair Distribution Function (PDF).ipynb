```python
'''Define rij:  rij is a matrix which results from the dot product of the fractional postion 
coordinates and the geometry information after it has been transformed into a matrix with the 
numbers only on the diagonal and all other places 0'''

#step 1.) Dr zhu code to help transform geometry to matrix

def para2matrix(cell_para):
        """ 1x6 (a, b, c, alpha, beta, gamma) -> 3x3 representation -> """
        matrix = np.zeros([3,3])
        matrix[0][0] = cell_para[0]
        matrix[1][0] = cell_para[1]*cos(cell_para[5])
        matrix[1][1] = cell_para[1]*sin(cell_para[5])
        matrix[2][0] = cell_para[2]*cos(cell_para[4])
        matrix[2][1] = cell_para[2]*cos(cell_para[3])*sin(cell_para[4])
        matrix[2][2] = sqrt(cell_para[2]**2 - matrix[2][0]**2 - matrix[2][1]**2)
        
        return matrix

#for i in part:
    para2matrix(i)
    
#x= np.array
#



```
