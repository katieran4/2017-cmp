```python
#HW b) Rewrite the Nuclear Decay Code Non-uniform random sampling
import numpy as np
import matplotlib.pyplot as plt


# Initials
NTl = 1000
NPb = 0
tau = 3.053*60
mu = np.log(2)/tau
dt = 1 
p = 1 - 2**(-dt/tau)
tmax = 1000
N=1000

t_points = np.arange(0.0, tmax, dt)
Tl = []
Pb = []


for t in t_points:
    Tl.append(NTl)
    Pb.append(NPb)
    
    #Find decaying atoms using non-uniform random numbers
    count = 0
    for i in range(NTl):
        z = np.random.random()
        rannum=(-np.log(1-z)/mu)
        Prob=rannum/NTl
        if Prob<p:
            count += 1
    NTl -= count
    NPb += count

plt.plot(t_points, Tl, label='Tl' , color='blue')
plt.plot(t_points, Pb, label='Pb' , color= 'darkorchid')
plt.xlabel("Time")
plt.ylabel("Number of atoms")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#define parameters

tau = 3.053*60
tmax=1000
N=1000
prob = np.log(2)/tau
NTl0=1000


Tl= NTl0 * np.ones([tmax-1])
Pb = np.zeros([tmax-1])
distr= []

for i in range(NTl0):
    z0 = np.random.random()
    rannum= int(-np.log(1-z0)/prob)
    #print(rannum)
    distr.append(rannum)
    if rannum < tmax:
        Tl[rannum:] -= 1
        Pb[rannum:] += 1
        
        
        
plt.plot(Tl, label='Tl' , color='dodgerblue')
plt.plot(Pb, label='Pb', color='darkviolet')
plt.title('Decay of Thallium(Tl) to Lead(Pb)')
plt.xlabel("Time (s)")
plt.ylabel("Number of atoms")

plt.legend()
plt.show()

plt.hist(distr)
plt.title('Distribution of Random Numbers Generated')
plt.xlim(0,2000)
plt.ylim(0,1000)
plt.show()
```
