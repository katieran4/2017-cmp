```python
from sklearn.datasets import load_iris
dataset = load_iris()
X = dataset.data
Y = dataset.target
print(X)
print(Y)
```

```python


from collections import OrderedDict
import pandas as pd

Iris_data = OrderedDict({'Sepal Length ': X[:,0],
             'Sepal Width': X[:,1],
             'Petal Length': X[:,2],
             'Petal Width': X[:,3],
             'Label': Y})

df = pd.DataFrame(Iris_data)

df
```

```python
import matplotlib.pyplot as plt

label = ['Sepal Length','Sepal Width','Petal Length','Petal Width']


fig, axarr = plt.subplots(4, 4)
fig.set_size_inches(18, 18)

for i in range(4):
    x1 = X[:,i]
    for j in range(4):
        x2 = X[:,j]
        if i == j:
            axarr[i,j].text(0.2, 0.5, label[i],fontsize=20)
        else:
            axarr[i,j].scatter(x1, x2, c=Y, cmap=plt.cm.nipy_spectral,edgecolor='k')
            
plt.show()
```

```python
'''Develop One Rule Algorithm'''

'''First step: transform all X data into  integers by taking the mean of
     each column and assigning all numbers above it as 1, below it as 0'''

import numpy as np

X_d = np.array( X>X.mean(), dtype='int')

from collections import OrderedDict
import pandas as pd

Iris_data_d = OrderedDict({'Sepal Length': X_d[:,0],
             'Sepal Width': X_d[:,1],
             'Petal Length': X_d[:,2],
             'Petal Width': X_d[:,3],
             'Label': Y})

df = pd.DataFrame(Iris_data_d)
df

```

```python
'''Rules\z!!'''

#SL--IF x=1, then y=0
countSL = 0
for itemSL in X_d[:,0]:
    if itemSL == 1:
        countSL += 1

#SW--IF x=0, then y=1
countSW = 0
for item in X_d[:,1]:
    if item == 0:
        countSW += 1
        
#PL--IF x=0, then y=0 ; IF x=1, then Y=2
countPL0 = 0
for itemPL0 in X_d[:,2]:
    if itemPL0 == 0:
        countPL0 += 1

countPL2 = 0
for itemPL2 in X_d[:,2]:
    if itemPL2 == 1:
        countPL2 += 1
        
        
#PW-- IF X=0, then Y=2
countPW = 0
for itemPW in X_d[:,3]:
    if itemPW == 0:
        countPW += 1
        
```

```python
'''from Sepal Length, number of Y=0 assigned and the error'''

Y0_SL= countSL

errorSL = (freqSL[1,1]+freqSL[2,1])/len(X_d) 

print (Y0_SL)
print (errorSL)

```

```python
'''from Sepal Width, number of Y=1 assigned and the error'''

Y1_SW=countSW

errorSW = (freqSW[0,0]+freqSW[2,0])/len(X_d) 

print (Y1_SW)
print (errorSW)
```

```python
'''from Petal Length, number of Y=0 & Y=2 assigned and the error'''

Y0_PL= countPL0 

Y2_PL= countPL2

errorPL0 = (freqPL[1,0])/len(X_d) 

errorPL2 = (freqPL[1,1])/len(X_d) 

print (Y0_PL)
print (errorPL0)

print (Y2_PL)
print (errorPL2)
```

```python
'''from Petal Width, number of Y=2 assigned and the error'''

Y2_PW=countPW

errorPW = (freqPW[0]+freqPW[1])/len(X_d) 

print (Y2_PW)
print (errorPW)
```
